# Feature Engineering 002 - Groupby Day Aggregations

```python
import pandas as pd
train = pd.read_parquet('../../data/train_FE001.parquet')
test = pd.read_parquet('../../data/test_FE001.parquet')
%matplotlib inline
```

```python
train.head()
```

```python
train['year_day'] = train['DT'].dt.year.astype('str') + '_' + train['DT'].dt.dayofyear.astype('str')
test['year_day'] = test['DT'].dt.year.astype('str') + '_' + test['DT'].dt.dayofyear.astype('str')
```

```python
train['Trans_Count_Day'] = train.groupby('year_day')['TransactionDT'].transform('count')
test['Trans_Count_Day'] = test.groupby('year_day')['TransactionDT'].transform('count')
```

```python
train['Mean_TransAmt_Day'] = train.groupby('year_day')['TransactionAmt'].transform('mean')
test['Mean_TransAmt_Day'] = test.groupby('year_day')['TransactionAmt'].transform('mean')
train['Sum_TransAmt_Day'] = train.groupby('year_day')['TransactionAmt'].transform('sum')
test['Sum_TransAmt_Day'] = test.groupby('year_day')['TransactionAmt'].transform('sum')
```

```python
train.to_parquet('../../data/train_FE002.parquet')
test.to_parquet('../../data/test_FE002.parquet')
```
