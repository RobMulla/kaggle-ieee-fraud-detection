# Addr1 first digit

```python
import pandas as pd
import matplotlib.pylab as plt
%matplotlib inline
```

```python
tr008 = pd.read_parquet('../../data/train_FE008.parquet')
te008 = pd.read_parquet('../../data/test_FE008.parquet')
```

```python
tr008['addr1'].unique()
```

```python
tr008['addr1'].dtypes
```

```python
tr008['addr1_first_digit'] = tr008['addr1'].astype(str).str[0].replace('n',-1).astype(int)
tr008['addr1_second_digit'] = tr008['addr1'].astype(str).str[1].replace('a',-1).astype(int)
tr008['addr1_third_digit'] = tr008['addr1'].astype(str).str[2].replace('n',-1).astype(int)
te008['addr1_first_digit'] = te008['addr1'].astype(str).str[0].replace('n',-1).astype(int)
te008['addr1_second_digit'] = te008['addr1'].astype(str).str[1].replace('a',-1).astype(int)
te008['addr1_third_digit'] = te008['addr1'].astype(str).str[2].replace('n',-1).astype(int)

tr008['addr2_first_digit'] = tr008['addr2'].astype(str).str[0].replace('n',-1).astype(int)
tr008['addr2_second_digit'] = tr008['addr2'].astype(str).str[1].replace('a',-1).astype(int)
tr008['addr2_third_digit'] = tr008['addr2'].astype(str).str[2].replace('n',-1).replace('.',-2).astype(int)
te008['addr2_first_digit'] = te008['addr2'].astype(str).str[0].replace('n',-1).astype(int)
te008['addr2_second_digit'] = te008['addr2'].astype(str).str[1].replace('a',-1).astype(int)
te008['addr2_third_digit'] = te008['addr2'].astype(str).str[2].replace('n',-1).replace('.',-2).astype(int)
```

```python
tr008['maxDT'].plot(kind='hist', figsize=(15, 5), alpha=0.5, label='Train', bins=100, title='maxDT')
te008['maxDT'].plot(kind='hist', figsize=(15, 5), alpha=0.5, label='Test', bins=100)
plt.legend()
```

```python
tr008['minDT'].plot(kind='hist', figsize=(15, 5), alpha=0.5, label='Train', bins=100, title='maxDT')
te008['minDT'].plot(kind='hist', figsize=(15, 5), alpha=0.5, label='Test', bins=100)
plt.legend()
```

```python
def plot_dist_train_test(train, test, feature, bins=100):
    train[feature].plot(kind='hist', figsize=(15, 5), alpha=0.5, label='Train', bins=bins, title='maxDT')
    test[feature].plot(kind='hist', figsize=(15, 5), alpha=0.5, label='Test', bins=bins)
    plt.legend()
```

```python
plot_dist_train_test(tr008, te008, 'addr1_first_digit', bins=10)
```

```python
plot_dist_train_test(tr008, te008, 'addr1_second_digit', bins=10)
```
