# Find Overlapping Columns

```python
train.columns
```

```python
train_fe002.head()
```

```python
train = train.set_index('TransactionID')
test = test.set_index('TransactionID')
```

```python
new_cols = [c for c in train.columns if c not in train_fe002.columns]
```

```python
new_cols
```

```python
train_merged = pd.concat([train_fe002, train[new_cols]], sort=False, axis=1)
test_merged = pd.concat([test_fe002, test[new_cols]], sort=False, axis=1)
```

```python
train_merged['D1_scaled'] = train_merged['D1_scaled'].astype('float32')
train_merged['D2_scaled'] = train_merged['D2_scaled'].astype('float32')
```

```python
test_merged['D1_scaled'] = test_merged['D1_scaled'].astype('float32')
test_merged['D2_scaled'] = test_merged['D2_scaled'].astype('float32')
```

```python
train_merged.to_parquet('../../data/train_FE003.parquet')
test_merged.to_parquet('../../data/test_FE003.parquet')
```
